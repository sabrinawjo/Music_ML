```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("../../Resources/archive/data.csv")
df.head()
```

```python

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['release_date']).year
df.head()
df = df.drop(['id','artists','name','release_date'], axis=1)

# df = df.loc[df['year'] != 1920 ]
# df = df.loc[df['year'] != 1921 ]
df.count()
```

```python
df.head()
```

```python
df.to_csv("songs_ml.csv")
```

# Select your features (columns)

```python
column_names = df.columns
column_names = column_names.tolist()
column_names.remove('explicit')
# column_names.remove('id')
# column_names.remove('name')
# column_names.remove('release_date')
# column_names.remove('artists')
column_names
```

```python
y = df['explicit']
y
```

```python
# Set features. This will also be used as your x values.
selected_features = df[column_names]
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(selected_features, y, random_state=18)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

# Create a StandardScater model and fit it to the training data

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

```python
sorted(zip(rf.feature_importances_, column_names), reverse=True)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [250, 300, 350],
              'max_depth': [125, 150, 175]}
grid = GridSearchCV(rf, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Training score:
grid.score(X_train_scaled, y_train)
```

```python
# Testing score:
grid.score(X_test_scaled, y_test)
```

```python
# Make prediction and save to variable for report.
predictions = grid.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
