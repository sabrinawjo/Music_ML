```python
import pandas as pd
```

```python
needGenre = pd.read_csv("../Resources/AprilUpdate/tracks_NEEDGenre.csv")
needGenre
```

```python
needGenre.columns
```

```python
needGenre["loudness_scale"] = 1-(needGenre["loudness"]/needGenre["loudness"].min())
needGenre["tempo_scale"] = needGenre["tempo"]/needGenre["tempo"].max()
needGenre["duration_scale"] = needGenre["duration_ms"]/needGenre["duration_ms"].max()
needGenre["popularity_scale"] = needGenre["popularity"]/needGenre["popularity"].max()
needGenre.head()
```

```python
needGenre.to_csv("../Resources/MLResources/needGenre_preML.csv")
```
