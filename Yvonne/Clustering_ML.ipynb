```python
import pandas as pd
```

```python
df = pd.read_csv("../Resources/AprilUpdate/tracks_genrePotential.csv")
df.head()
```

```python
df = df[df["loudness"]<=0]
df.head()
```

```python
df["loudness_scale"] = 1-(df["loudness"]/df["loudness"].min())
df["tempo_scale"] = df["tempo"]/df["tempo"].max()
df["duration_scale"] = df["duration_ms"]/df["duration_ms"].max()
df["popularity_scale"] = df["popularity"]/df["popularity"].max()
df.head()
```

```python
df.columns
```

```python
df.to_csv("../Resources/clean_clustering_ml_data.csv", index=False)
```
