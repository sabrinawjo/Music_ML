```python
song_2000 = "Spotify-2000.csv"
song_2000_df = pd.read_csv(song_2000)
song_2000_df = song_2000_df.drop(columns = ["Index"])
song_2000_df.head()
```

```python
song_2000_df = song_2000_df.dropna()
song_2000_df["Title"].count()
```

```python
genre = song_2000_df.drop("Top Genre", axis=1)
feature_names = song_2000_df.columns
```

```python
genre.head()
```

```python
target = song_2000_df["Top Genre"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(genre, target, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
```
